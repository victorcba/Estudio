```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
image_path = 'imagen.jpg'
image = cv2.imread(image_path, 0)  # Leer la imagen en escala de grises

# Obtener el perfil de intensidad
profile_x = np.mean(image, axis=0)  # Perfil de intensidad a lo largo del eje x
profile_y = np.mean(image, axis=1)  # Perfil de intensidad a lo largo del eje y

# Graficar el perfil de intensidad
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(profile_x)
plt.title('Perfil de Intensidad - Eje X')
plt.xlabel('Posicin en el eje X')
plt.ylabel('Intensidad')
plt.subplot(1, 2, 2)
plt.plot(profile_y)
plt.title('Perfil de Intensidad - Eje Y')
plt.xlabel('Posicin en el eje Y')
plt.ylabel('Intensidad')
plt.tight_layout()
plt.show()

# Resaltar las lneas de los perfiles de intensidad en la imagen original
image_with_lines = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
line_color = (0, 0, 255)  # Rojo (BGR)
line_thickness = 2

# Resaltar el perfil de intensidad en el eje X
x_start = 0
x_end = image.shape[1]
x_line_position = np.argmax(profile_x)  # Posicin con mayor intensidad en el perfil X
cv2.line(image_with_lines, (x_line_position, x_start), (x_line_position, x_end), line_color, line_thickness)

# Resaltar el perfil de intensidad en el eje Y
y_start = 0
y_end = image.shape[0]
y_line_position = np.argmax(profile_y)  # Posicin con mayor intensidad en el perfil Y
cv2.line(image_with_lines, (y_start, y_line_position), (y_end, y_line_position), line_color, line_thickness)

# Mostrar la imagen con las lneas resaltadas
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_with_lines, cv2.COLOR_BGR2RGB))
plt.title('Imagen con Perfiles de Intensidad')
plt.tight_layout()
plt.show()
```
