```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
image_path = 'imagen.jpg'
image = cv2.imread(image_path)

# Definir el kernel del filtro pasa bajo (filtro de promedio)
kernel_size = (5, 5)  # Tamaño del kernel del filtro
kernel = np.ones(kernel_size, dtype=np.float32) / (kernel_size[0] * kernel_size[1])

# Aplicar el filtro pasa bajo mediante convolución
filtered_image = cv2.filter2D(image, -1, kernel)

# Mostrar la imagen original y la imagen filtrada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Filtrada (Pasa Bajo)')
plt.tight_layout()
plt.show()
```
